```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
def mapWithPointLists(lon,lat, image):
    fig, ax = plt.subplots(figsize=(10,20))
    m = Basemap(projection='merc',lat_0=(northlimit+southlimit)/2,
                lon_0=(westlimit+eastlimit)/2,
                llcrnrlon=westlimit, llcrnrlat=southlimit,
                urcrnrlon=eastlimit, urcrnrlat=northlimit,
                resolution='l')

    X,Y = m(lon,lat)
    m.drawcoastlines()
    m.drawstates()
    m.drawcountries()
    m.drawmapboundary(fill_color='#46bcec')
    m.fillcontinents(color='#94e27f', lake_color='#46bcec')

    #ax.scatter(X,Y, zorder=5)
    for i, (x, y) in enumerate(zip(X, Y), start=1):
        #ax.annotate(str(i), (x,y), xytext=(5, 5), textcoords='offset points')
        imscatter(x,y,image,ax,zoom=0.16)
        
def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
```

```python
num = 15
lat = 5.5 * np.random.random(num) + 47
lon = 6.5 * np.random.random(num) + 6.5

incubLat = [53.57632,51.96236,52.52437]
incubLon = [10.01534,7.62571, 13.41053]

secondLat = [48.401082,49.581085,51.339695,51.045925,52.030228,52.480909]
secondLon = [9.987608, 9.362160, 12.373075,7.019220 , 8.532471,10.550783]
secondLat2 = incubLat + secondLat
secondLon2 = incubLon + secondLon

image_path = "C:\\Users\\Philipp\\Downloads\\v7_We Are Human.png"
mapWithPointLists(secondLon2,secondLat2, image_path)

plt.show()
```
